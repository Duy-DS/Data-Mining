### Tải dữ liệu vào bộ nhớ: Xem và quản lý dễ dàng bằng Pandas
Đầu tiên, chúng ta cần tải dữ liệu vào bộ nhớ để Python có thể tương tác với nó. Pandas sẽ là thư viện quản lý và thao tác dữ liệu của chúng ta.

```python
# load data into Pandas
import pandas as pd
df = pd.read_csv("F:/Study Materials/Môn Ngành/Data Mining - PY/Python-Data-Mining-Quick-Start-Guide-master/data/iris.csv")
```

Hãy sử dụng một số tính năng tích hợp sẵn của Pandas để kiểm tra tính hợp lệ của dữ liệu tải lên và đảm bảo mọi thứ đã được tải đúng cách. ### Vẽ biểu đồ và khám phá dữ liệu: Khai thác sức mạnh của Seaborn
Bây giờ, hãy bắt đầu phân tích của chúng ta với chương trình vẽ đồ thị đóng gói sẵn của Seaborn, được gọi là **pairplot**, để trực quan hóa các mối quan hệ đặc trưng từng cặp. Bạn có thể sử dụng chương trình này để tìm kiếm các mối quan hệ, ứng viên cho các nhóm, các giá trị ngoại lai có thể có và trực giác về các chiến lược hạ nguồn cần nghiên cứu để phân tích. Mỗi ô ngoài đường chéo là một biểu đồ phân tán từng cặp và các đường chéo được điền các phân phối đơn biến.

Đôi khi, biểu đồ histogram dễ hiểu hơn biểu đồ mật độ xác suất trong việc hiểu phân phối. Lúc này, chúng ta có thể chọn hai biến và biểu diễn chúng trên biểu đồ phân tán bằng **lmplot** của Seaborn. Nếu tập dữ liệu của bạn có nhiều hơn năm đặc trưng, ​​các mối quan hệ biến quan trọng có thể không được hiển thị trên cùng một cửa sổ của biểu đồ cặp. Khái niệm này nghe có vẻ phức tạp, nhưng chỉ cần nhìn vào biểu đồ là bạn sẽ dễ dàng hiểu được ý tưởng, và đó chính là ý tưởng. Bạn càng thấy nhiều biểu đồ violin, bạn sẽ càng yêu thích chúng.

LƯU Ý: Theo mặc định, Seaborn sẽ thêm trung vị và khoảng tứ phân vị (50% giữa) vào mỗi violin trong biểu đồ. Bạn có thể thay đổi điều này bằng cách sử dụng đối số **bên trong**. Điều này được giải thích trong tài liệu trực tuyến của Seaborn về biểu đồ violin.

```python
sns.violinplot(x='species',y='petal length in cm', data=df)
```

### Biến đổi Dữ liệu: PCA và LDA với Scikit-learn
Thông thường, một phép biến đổi có thể giúp dữ liệu dễ hiểu hơn. Cụ thể, các nhà khoa học dữ liệu sử dụng phép biến đổi để xoay dữ liệu quanh trục có biến thiên tổng thể hoặc quan trọng nhất, nhằm mục đích biểu diễn thông tin tương tự trong số chiều ít hơn. Chúng ta có thể sử dụng tập dữ liệu mống mắt làm ví dụ để lấy bốn đặc trưng và biểu diễn thông tin tương tự trong hai chiều. Hãy bắt đầu với **phân tích thành phần chính (PCA)** để định hướng dữ liệu vào các trục có biến thiên cao nhất. Tập mống mắt chỉ có bốn chiều, nhưng kỹ thuật này có thể được sử dụng trên dữ liệu có hàng chục hoặc hàng trăm đặc trưng.

```python
# reduce dimensions with PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
out_pca = pca.fit_transform(df[['sepal length in cm',
                                'sepal width in cm',
                                'petal length in cm',
                                'petal width in cm']])
```

Bây giờ chúng ta hãy tạo một khung dữ liệu pandas với dữ liệu đầu ra và sử dụng kiểm tra tính hợp lý **head()** để xem chúng ta có gì:

```python
df_pca = pd.DataFrame(data = out_pca, columns = ['pca1', 'pca2'])
print(df_pca.head())
```

Trông có vẻ ổn, ngoại trừ việc chúng ta thiếu cột mục tiêu hoặc nhãn ("species"). Hãy thêm cột bằng cách nối với khung dữ liệu gốc. Điều này sẽ cho chúng ta một khung dữ liệu pca ("df_pca") sẵn sàng cho các tác vụ xử lý và dự đoán tiếp theo. Sau đó, hãy vẽ biểu đồ và xem dữ liệu đã chuyển đổi của chúng ta trông như thế nào khi chỉ vẽ trên hai chiều.

```python
df_pca = pd.concat([df_pca, df[['species']]], axis = 1)
print(df_pca.head())
sns.lmplot(x="pca1", y="pca2", hue="species", data=df_pca, fit_reg=False)
```

Bây giờ chúng ta đã có dữ liệu đa chiều được biểu diễn dưới dạng hai chiều dễ hiểu và dễ vẽ đồ thị. Nhưng liệu chúng ta có thể làm tốt hơn không? Mục tiêu của PCA là định hướng dữ liệu theo hướng biến thiên lớn nhất. Tuy nhiên, nó bỏ qua một số thông tin quan trọng từ tập dữ liệu của chúng ta. Cụ thể là, các nhãn không được sử dụng. Có lẽ chúng ta có thể trích xuất các vectơ biến đổi tốt hơn nữa nếu bao gồm các nhãn. Kỹ thuật giảm chiều có nhãn phổ biến nhất được gọi là **phân tích phân biệt tuyến tính (LDA)**. Phép toán bên dưới sẽ nhóm theo nhãn lớp, sau đó tìm hướng phân tách lớn nhất giữa các lớp.

LƯU Ý: Việc bỏ qua nhãn trong bước biến đổi có thể là mong muốn đối với một số câu lệnh bài toán, đặc biệt là những câu lệnh có nhãn lớp không đáng tin cậy, để tránh kéo các vectơ thành phần đã rút gọn theo hướng không hữu ích. Vì lý do này, tôi khuyên bạn nên luôn bắt đầu với PCA trước khi quyết định xem tôi có cần làm thêm hay không. Thực tế, trừ khi tập dữ liệu của bạn lớn, thời gian tính toán cho PCA khá ngắn, vì vậy việc bắt đầu từ đây cũng không có hại gì.

```python
# reduce dimensions with LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis(n_components=2)

# format dataframe
out_lda = lda.fit_transform(X=df.iloc[:,:4], y=df['species'])
df_lda = pd.DataFrame(data = out_lda, columns = ['lda1', 'lda2'])
df_lda = pd.concat([df_lda, df[['species']]], axis = 1)

# sanity check
print(df_lda.head())

# plot
sns.lmplot(x="lda1", y="lda2", hue="species", data=df_lda, fit_reg=False)
```

Biểu đồ phân tán có thể khiến bạn nghĩ rằng các kỹ thuật PCA và LDA đã thực hiện cùng một phép biến đổi trên dữ liệu. Hãy cùng xem xét kỹ hơn thành phần đầu tiên của mỗi kỹ thuật bằng cách sử dụng thủ thuật biểu đồ violin mạnh mẽ. Trước tiên, chúng ta sẽ bắt đầu với PCA:

```python
sns.violinplot(x='species',y='pca1', data=df_pca).set_title("Violin plot: Feature = PCA_1")
```

Bây giờ chúng ta hãy vẽ biểu đồ thành phần LDA đầu tiên:

```python
sns.violinplot(x='species',y='lda1', data=df_lda).set_title("Violin plot: Feature = LDA_1")
```

### Định lượng Phân tách: Phân cụm K-means và Điểm Silhouette
Phân tách lớp khó nhất trong tập dữ liệu này là "versicolor" và "virginica". Các violon cho mỗi lớp này cho thấy hai kỹ thuật thực sự tạo ra các kết quả khác nhau. Sử dụng phân phối "setosa" làm tham chiếu trong cả hai biểu đồ, phân phối LDA versicolor chặt chẽ hơn (rộng hơn và ngắn hơn) so với phân phối PCA, khiến khoảng tứ phân vị của nó tách biệt hơn so với khoảng tứ phân vị của phân phối virginica. Nếu phân tích này chưa đủ chặt chẽ, chúng ta có thể dễ dàng định lượng sự khác biệt này bằng cách sử dụng thuật toán phân cụm trên dữ liệu. Hãy sử dụng thuật toán phân cụm **k-means** để nhóm dữ liệu lại với nhau về mặt toán học, sau đó sử dụng thước đo định lượng được gọi là **hệ số bóng** để chấm điểm độ chặt chẽ của các cụm kết quả. Điểm cao hơn đồng nghĩa với các cụm chặt chẽ hơn. Vì thuật toán k-means rất đơn giản và chất lượng nhóm có liên quan trực tiếp đến chất lượng dữ liệu đầu vào nên các cụm chặt chẽ hơn chứng tỏ rằng các đặc điểm đầu vào phân tách các lớp tốt hơn.

```python
# cluster With k-means and check silhouette score
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# initialize k-means algo object
kmns = KMeans(n_clusters=3, random_state=42)

# fit algo to pca and find silhouette score
out_kms_pca = kmns.fit_predict(out_pca)
silhouette = silhouette_score(out_pca, out_kms_pca)
print("PCA silhouette score = " + str(silhouette))

# fit algo to lda and find silhouette score
out_kms_lda = kmns.fit_predict(out_lda)
silhouette = silhouette_score(out_lda, out_kms_lda)
print("LDA silhouette score = %2f " % silhouette)
```

Kết quả đầu ra ở trên cho thấy các lớp LDA được phân tách tốt hơn. Điều này hợp lý vì hàm LDA có nhiều thông tin hơn, cụ thể là các lớp cần phân tách.

### Ra Quyết định hoặc Dự đoán
Trước khi xây dựng dự đoán, chúng ta cần phân tách dữ liệu thành tập huấn luyện và tập kiểm tra. Kiểm định mô hình là một chủ đề lớn và rất quan trọng sẽ được đề cập ở phần sau của cuốn sách, nhưng với mục đích của ví dụ đầu cuối này, chúng ta sẽ thực hiện phân tách huấn luyện-kiểm tra cơ bản. Sau đó, chúng ta sẽ xây dựng mô hình dự đoán trên dữ liệu huấn luyện và chấm điểm nó trên dữ liệu kiểm tra bằng điểm F1.

LƯU Ý: Tôi khuyên bạn nên sử dụng hạt giống ngẫu nhiên cho hầu hết các lựa chọn dữ liệu ngẫu nhiên. Hạt giống này cho bộ tạo số giả ngẫu nhiên biết nơi bắt đầu quy trình ngẫu nhiên của nó. Kết quả là cùng một lựa chọn ngẫu nhiên mỗi lần. Trong ví dụ này, tôi đã sử dụng hạt giống ngẫu nhiên khi phân tách thành tập huấn luyện và tập kiểm tra. Bây giờ, nếu tôi ngừng làm việc trên dự án và tiếp tục sau đó, tôi có thể phân tách bằng hạt giống ngẫu nhiên và nhận được chính xác cùng một tập huấn luyện và tập kiểm tra. Tôi đã sử dụng "42" cho hạt giống của mình vì nó phổ biến trong lĩnh vực này do sự phổ biến của bộ truyện "Cẩm nang du lịch thiên hà" của Douglas Adams.

```python
# Split into train/validation/test set
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(df_lda, test_size=0.3, random_state=42)

# Sanity check
print('train set shape = ' + str(df_train.shape))
print('test set shape = ' + str(df_test.shape))
print(df_train.head())
```

Bây giờ chúng ta có thể chuyển sang phần dự đoán. Trước tiên, hãy thử **Máy vectơ hỗ trợ (SVM)** bằng mô-đun **SVC**. Lưu ý cách các đối tượng phân loại trong Scikit-learn có các lệnh gọi API tương tự với các phép biến đổi PCA và LDA đã đề cập trước đó. Vì vậy, một khi bạn đã nắm được ý chính của thư viện, bạn có thể học cách áp dụng các phép biến đổi, bộ phân loại hoặc các phương pháp khác nhau mà không tốn nhiều công sức.

```python
# classify with SVM
from sklearn.svm import SVC
from sklearn.metrics import f1_score
clf = SVC(kernel='rbf', C=0.8, gamma=10)
clf.fit(df_train[['lda1', 'lda2']], df_train['species'])

# predict on test set
y_pred = clf.predict(df_test[['lda1', 'lda2']])
f1 = f1_score(df_test['species'], y_pred, average='weighted')

# check prediction score
print("f1 score for SVM classifier = %2f " % f1)
```

Điểm F1 cho bộ phân loại này là 0,79 được tính toán trên tập kiểm tra. Tại thời điểm này, chúng ta có thể thử thay đổi thiết lập mô hình và điều chỉnh lại. Tham số C được đặt thành 0,8 trong lần chạy đầu tiên, sử dụng đối số "C=0,8" trong khởi tạo đối tượng "clf". C là một giá trị phạt và được gọi là **siêu tham số**, nghĩa là nó là một thiết lập mà nhà phân tích có thể sử dụng để điều chỉnh việc điều chỉnh theo một hướng nhất định. Ở đây, chúng ta sẽ sử dụng siêu tham số phạt C để điều chỉnh mô hình theo hướng dự đoán tốt hơn. Hãy thay đổi nó từ 0,8 thành 1, về cơ bản là tăng giá trị phạt.

LƯU Ý: C là giá trị phạt trong SVM. Nó kiểm soát độ lớn của giá trị phạt đối với các ví dụ bị phân loại sai trong quá trình điều chỉnh mô hình. Để dễ hiểu hơn, nó được gọi là "giá trị phạt biên mềm" vì nó điều chỉnh độ cứng/mềm của đường phân tách kết quả được vẽ. Các siêu tham số phổ biến cho máy vectơ hỗ trợ sẽ được đề cập chi tiết hơn trong chương sau.

```python
# classify with SVM
from sklearn.svm import SVC
from sklearn.metrics import f1_score
clf = SVC(kernel='rbf', C=1, gamma=10)
clf.fit(df_train[['lda1', 'lda2']], df_train['species'])
y_pred = clf.predict(df_test[['lda1', 'lda2']])
f1 = f1_score(df_test['species'], y_pred, average='weighted')
print("f1 score for SVM classifier = %2f " % f1)
```

Điểm F1 cho bộ phân loại này hiện là 0,85. Bước tiếp theo hiển nhiên là điều chỉnh các tham số và tối đa hóa điểm F1. Tất nhiên, việc thay đổi tham số, tinh chỉnh, phân tích, lặp lại, v.v. sẽ rất tốn công sức. Thay vào đó, bạn có thể sử dụng **tìm kiếm lưới** để tự động hóa việc tham số hóa này. Tìm kiếm lưới và **xác thực chéo** sẽ được trình bày chi tiết trong các chương sau. Một phương pháp thay thế cho việc sử dụng tìm kiếm lưới là chọn một thuật toán không cần điều chỉnh. Một thuật toán phổ biến gần như không cần điều chỉnh là **Rừng ngẫu nhiên**. "Rừng" ám chỉ cách phương pháp này cộng nhiều cây quyết định lại với nhau thành một dự đoán đã được bỏ phiếu.

```python
# classify with RF
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=2, random_state=42)
clf.fit(df_train[['lda1', 'lda2']], df_train['species'])
y_pred = clf.predict(df_test[['lda1', 'lda2']])
f1 = f1_score(df_test['species'], y_pred, average='weighted')

# check prediction score
print("f1 score for SVM classifier = %2f " % f1)
```
